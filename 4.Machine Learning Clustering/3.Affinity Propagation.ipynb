```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import AffinityPropagation
import seaborn as sns
```

```python
# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')
dataset
```

```python
# Select features: Annual Income and Spending Score
X = dataset.iloc[:, 3:5].values
X
```

```python
# Apply Affinity Propagation
ap_model = AffinityPropagation(damping=0.9, random_state=42)
label = ap_model.fit_predict(X)
```

```python
# Add cluster labels to the original dataset
supervised = pd.DataFrame(dataset)
supervised['Cluster_group'] = label
```

```python
# Save clustered data
supervised.to_csv("affinity_clusters.csv", index=False)
```

```python
# Visualize clusters
sns.lmplot(data=supervised, 
           x=supervised.columns[3], 
           y=supervised.columns[4], 
           hue='Cluster_group', 
           fit_reg=False, 
           legend=True, 
           legend_out=True)
plt.title('Affinity Propagation Clusters')
plt.show()
```
