```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import SpectralClustering
import seaborn as sns
```

```python
# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
# Select features: Annual Income and Spending Score
X = dataset.iloc[:, 3:5].values
```

```python
# Apply Spectral Clustering
sc_model = SpectralClustering(n_clusters=5, affinity='nearest_neighbors', assign_labels='kmeans', random_state=42)
label = sc_model.fit_predict(X)
from sklearn.metrics import silhouette_score
print("Silhouette Score:", silhouette_score(X, label))
```

```python
# Add cluster labels to the original dataset
supervised = pd.DataFrame(dataset)
supervised['Cluster_group'] = label
supervised
```

```python
# Save clustered data
supervised.to_csv("spectral_clusters.csv", index=False)
```

```python
# Visualize clusters
sns.lmplot(data=supervised, 
           x=supervised.columns[3], 
           y=supervised.columns[4], 
           hue='Cluster_group', 
           fit_reg=False, 
           legend=True, 
           legend_out=True)

plt.title('Spectral Clustering')
plt.show()
```

```python
from sklearn.metrics.pairwise import rbf_kernel
from scipy.linalg import eigh

# Compute affinity matrix
affinity_matrix = rbf_kernel(X, gamma=0.1)

# Compute Laplacian
D = np.diag(affinity_matrix.sum(axis=1))
L = D - affinity_matrix

# Compute eigenvalues
eigenvalues, _ = eigh(L)

plt.plot(sorted(eigenvalues))
plt.title('Spectral Clustering â€“ Eigenvalue Spectrum')
plt.xlabel('Index')
plt.ylabel('Eigenvalue')
plt.show()
```
