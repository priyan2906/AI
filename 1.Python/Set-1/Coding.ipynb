```python
#4
def factorial(n):
    result = 1
    for i in range(1, n+1):
        result *= i
    return result
print(factorial(5))
```

```python
#5
def isPrime(n):
    if n<2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n%i == 0:
            return False
    return True
num = 7
if isPrime(7):
    print(f"The given number {num} is a prime number")
else:
    print(f"The given number {num} is not a prime number")
```
