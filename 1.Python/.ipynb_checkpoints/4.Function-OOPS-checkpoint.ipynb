```python
#if-elif
age=int(input("Enter age:"))
```

```python
if(age<18):
    print("Children")
elif(age<35):
    print("Adult")
elif(age<59):
    print("Citizen")
else:
    print("Senior Citiizen")
```

```python
"""Single input with multiple conditions then only you need to select 'if family'
When to use:
If - Single condition
If-else - Two conditions
If-elif - More than two conditions"""
```

```python
#for-loop
lists=[15,35,47,65]
#list can hold multiple values in a single variable
#syntax
'''for temp in main variable:
        print()'''
```

```python
for temp in lists:
    print(temp)
```

```python
#You can use for loop with if family, when you have multiple conditions
lists=[15,34,47,65]
for age in lists:
    if(age<18):
        print("Children")
    elif(age<35):
        print("Adult")
    elif(age<59):
        print("Citizen")
    else:
        print("Senior Citizen")
```

```python
#Find odd or even
#Way 1
num=int(input("Enter the number"))
if((num%2)==0):
    print("Even number")
else:
    print("Odd number")
```

```python
#Way 2
num=int(input("Enter the number"))
if((num%2)==1):
    print("Odd number")
else:
    print("Even number")
```
