```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import Birch
import seaborn as sns

# Load dataset
dataset = pd.read_csv('Mall_Customers.csv')

# Select features: Annual Income and Spending Score
X = dataset.iloc[:, 3:5].values

# Apply BIRCH clustering
birch_model = Birch(n_clusters=5)
label = birch_model.fit_predict(X)

# Add cluster labels to the original dataset
supervised = pd.DataFrame(dataset)
supervised['Cluster_group'] = label

# Save clustered data
supervised.to_csv("birch_clusters.csv", index=False)

# Visualize clusters
sns.lmplot(data=supervised, 
           x=supervised.columns[3], 
           y=supervised.columns[4], 
           hue='Cluster_group', 
           fit_reg=False, 
           legend=True, 
           legend_out=True)

plt.title('BIRCH Clustering')
plt.show()
```
