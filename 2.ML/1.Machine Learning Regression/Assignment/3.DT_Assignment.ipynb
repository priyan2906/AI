```python
import pandas as pd
```

```python
dataset = pd.read_csv("insurance_pre.csv")
```

```python
dataset = pd.get_dummies(dataset,drop_first=True)
```

```python
independent=dataset[['age', 'bmi', 'children', 'sex_male', 'smoker_yes']]
```

```python
dependent=dataset['charges']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson',splitter='random')
regressor=regressor.fit(X_train,y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
# Measuring accuracy using R² score
from sklearn.metrics import r2_score
# Finding model accuracy using R² score
r_score = r2_score(y_test,y_pred)
#Display r_score value 0 to 1, higher value will be good prediction
r_score
```

```python
import pickle
filename="finalized_model_dt_assign.sav"
pickle.dump(regressor,open(filename,'wb'))
```
