```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,columns=['State'],drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend','State_Florida', 'State_New York']]
```

```python
dependent=dataset[["Profit"]]
```

```python
'''Splitting data into training set and testing set
   test_size = 0.30 means 30% of data goes for testing
   random_state ensures same split every time (for reproducibility)'''
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='poisson',splitter='random')
regressor=regressor.fit(X_train,y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
y_pred = regressor.predict(X_test)
```

```python
# Measuring accuracy using R score
from sklearn.metrics import r2_score
# Finding model accuracy using R score
r_score = r2_score(y_test,y_pred)
#Display r_score value 0 to 1, higher value will be good prediction
r_score
```
